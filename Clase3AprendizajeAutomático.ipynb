```python
import pandas as pd
#lectura del archivo de ingreso
pacientes=pd.read_csv('Pacientes2.csv',engine='python',index_col=0)
pacientes.head()
#obtener información del data set
pacientes.info()

#variables independientes
x=pacientes.iloc[:,1:11]
#variables dependientes
y=pacientes.iloc[:,0]

x.head()


#dividir datos de entrenamiento y testeo
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0, )

x_train.info()

from sklearn.tree import DecisionTreeClassifier

clasificador=DecisionTreeClassifier(criterion='entropy',random_state=0, max_depth=3)
#se entrena el modelo
clasificador.fit(x_train,y_train)

y_pred=clasificador.predict(x_test)

#matriz de confusión (ROC)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm

#Graficar el árbol
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
tree.plot_tree(clasificador,filled=True)
plt.show()
```
